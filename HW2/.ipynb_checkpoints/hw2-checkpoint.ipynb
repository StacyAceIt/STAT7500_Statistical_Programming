## Problem 1: What are these dates and why are they interesting?
    -67019
    9222
    -7427
    17834
    4859
Determine the dates for the above unformatted dates and try to determine why they are mildly interesting.

```sas
options center nodate nonumber ps=60 ls=72;
dm'log;clear;out;clear;';
```

```sas
data mydata;
input numbers; 
datalines;
    -67019
    9222
    -7427
    17834
    4859
;
run;

proc print data=mydata;
    title 'dates chart';
run;    
```

```sas
data date1;
    set mydata;
    format formatted_date ddmmyy10.;
    formatted_date = numbers;
run;

proc print data=date1;
    title 'Problem 1: Formatted Dates';
run;
```

#### Why are they interesting?
* The formatted dates show that SAS starts counting time from Jan 1, 1960. 
* Positive numbers represents dates after Jan 1, 1960. 
* Negative numbers represents dates before Jan 1, 1960. 
* 0 represents Jan 1, 1960.

### a) 	Create a data set containing only FICE, State, all average salaries and average compensations. Print the dataset.

### c) 	Within the subset AK created in b), create new variables which are:
####    i.  the difference between compensation and salary for full professors


```sas
data diff_fp;
    set from_AK(keep = C_fp S_fp FICE State);
    diff = C_fp - S_fp;
run;

data diff_fp2;
    set diff_fp(drop = C_fp S_fp);
run;
```

```sas
proc print data = diff_fp2;
    title 'difference between compensation C_fp and salary S_fp for fulltime professors';
run;
```

```sas
proc means data = diff_fp2 n mean median std p25 p75 qrange minimum maximum;
    var diff;
    title 'statistics of diff_fp2';
run;
```

## Problem 3: For this problem, use the corn.xls dataset which can be found on blackboard.  Here is a description of the dataset.
    There are three variables:

    year: from 1909 to 1927 

    yield: yearly corn yield in bushels per acre, in six Corn Belt
     states(Iowa, Illinois, Nebraska, Missouri, Indiana, and
     Ohio).

    rain: rainfall measurements in inches, in the six states.

    Read in the corn dataset using proc import.
    Create a new variable catyield to categorize the corn yield: 
    if yield < 32, then catyield = ‘poor’;
    if yield >= 32, then catyield = ‘good’.   Create a new variable catrain to categorize the amount of rainfall: 
    if rain < = 9.7, then catrain = ‘drought’;
    if 9.7 < rain < = 12, then catrain = ‘normal’;
    if rain >12, then catrain = ‘wet’.
    Print the new dataset.  Turn in your SAS code, log and output files.

```sas
proc import out = corn_dataset datafile = "/folders/myfolders/STAT7500_Statistical_Programming/HW2/corn.xls" dbms = xls replace;
    getnames = yes;
run;
```

```sas
proc print data = corn_dataset;
    title 'corn.xls dataset';
run;
```

```sas
data corn_data;
    set corn_dataset;
    if yield < 32 then catyield = 'poor';
    else catyield = 'good';
run;
```

```sas
proc print data = corn_data;
    title 'corn.xls dataset with catyield';
run;
```

```sas
data corn_data2;
    set corn_data;
    if rain <= 9.7 then catrain = 'drought';
    else if rain <= 12 then catrain = 'normal';
    else catrain = 'wet';
run;
```

```sas
proc print data = corn_data2;
    title 'corn.xls dataset with catyield and catrain';
run;
```
